```python
import copy
import math
import numpy

X = "X"
O = "O"
EMPTY = None
user = None
ai = None
def initial_state():
    """
    Returns starting state of the board.
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """
    Returns player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return ai
    return user

def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def get_horizontal_winner(board):
    # check horizontally
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
                break
        if winner_val:
            return winner_val
    return winner_val

def get_vertical_winner(board):
    # check vertically
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
                break
        if winner_val:
            return winner_val
    return winner_val

def get_diagonal_winner(board):
    # check diagonally
    winner_val = None
    board_len = len(board)
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
            break
    if winner_val:
        return winner_val

    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None
            break
    return winner_val

def winner(board):
    """
    Returns the winner of the game, if there is one.
    """
    winner_val = get_horizontal_winner(board) or get_vertical_winner(board) \
                 or get_diagonal_winner(board) or None
    return winner_val

def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) != None:
        return True
    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True

def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    """
    Returns the optimal action for the current player on the board.
    """
    current_player = player(board)
    if current_player == X:
        min_val = -math.inf
        move = None
        for action in actions(board):
            check = minValue(result(board, action))
            if check > min_val:
                min_val = check
                move = action
        return move
    else:
        max_val = math.inf
        move = None
        for action in actions(board):
            check = maxValue(result(board, action))
            if check < max_val:
                max_val = check
                move = action
        return move

if __name__ == "__main__":
    board = initial_state()
    ai_turn = False
    print("Choose a player")
    user = input()
    if user == "X":
        ai = "O"
    else:
        ai = "X"

    while True:
        game_over = terminal(board)
        playr = player(board)

        if game_over:
            winner_val = winner(board)
            if winner_val is None:
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {winner_val} wins.")
            break
        else:
            if user != playr and not game_over:
                if ai_turn:
                    move = minimax(board)
                    board = result(board, move)
                    ai_turn = False
                    print(numpy.array(board))
            elif user == playr and not game_over:
                ai_turn = True
                print("Enter the position to move (row,col)")
                try:
                    i = int(input("Row:"))
                    j = int(input("Col:"))
                    if 0 <= i < len(board) and 0 <= j < len(board[0]) and board[i][j] == EMPTY:
                        board = result(board, (i, j))
                        print(numpy.array(board))
                    else:
                        print("Invalid move. Try again.")
                        ai_turn = False # User needs to try again
                except ValueError:
                    print("Invalid input. Please enter numbers for row and column.")
                    ai_turn = False # User needs to try again

```

Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 3x3.

```python
import math

PLAYER = 'O'
AI = 'X'

# Kiểm tra có ai thắng không
def check_win(board):
    win_states = [
        [0,1,2], [3,4,5], [6,7,8],         # hàng
        [0,3,6], [1,4,7], [2,5,8],         # cột
        [0,4,8], [2,4,6]                   # chéo
    ]
    for a, b, c in win_states:
        if board[a] == board[b] == board[c] != ' ':
            return board[a]
    return None

def is_full(board):
    return ' ' not in board

# Thuật toán minimax
def minimax(board, is_maximizing):
    winner = check_win(board)

    # Trả điểm nếu có người thắng
    if winner == AI: return 10
    if winner == PLAYER: return -10
    if is_full(board): return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = AI
                score = minimax(board, False)
                board[i] = ' '
                best_score = max(best_score, score)
        return best_score
    else:
        best_score = math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = PLAYER
                score = minimax(board, True)
                board[i] = ' '
                best_score = min(best_score, score)
        return best_score

# Tìm nước đi tốt nhất cho máy
def best_move(board):
    best_score = -math.inf
    move = -1
    for i in range(9):
        if board[i] == ' ':
            board[i] = AI
            score = minimax(board, False)
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i
    return move

# -------------------------
# Demo
board = [
    'X', 'O', 'X',
    'O', 'X', ' ',
    ' ', ' ', 'O'
]

print("AI nên đánh vào vị trí:", best_move(board))
```

Cài đặt thuật toán cắt tỉa Alpha-beta với ứng dụng vào bài toán TicTacToe với ma
trận 3x3.

```python
import math

PLAYER = 'O'   # Người
AI = 'X'       # Máy

# Kiểm tra thắng
def check_win(board):
    win_states = [
        [0,1,2], [3,4,5], [6,7,8],
        [0,3,6], [1,4,7], [2,5,8],
        [0,4,8], [2,4,6]
    ]
    for a,b,c in win_states:
        if board[a] == board[b] == board[c] != ' ':
            return board[a]
    return None

def is_full(board):
    return ' ' not in board


# -------------------------------
#      MINIMAX + ALPHA-BETA
# -------------------------------
def alphabeta(board, depth, alpha, beta, is_maximizing):
    winner = check_win(board)

    # Trả điểm nếu có người thắng
    if winner == AI: return 10 - depth
    if winner == PLAYER: return depth - 10
    if is_full(board): return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = AI
                score = alphabeta(board, depth + 1, alpha, beta, False)
                board[i] = ' '

                best_score = max(best_score, score)
                alpha = max(alpha, best_score)

                if beta <= alpha:
                    break  # CẮT TỈA

        return best_score

    else:
        best_score = math.inf
        for i in range(9):
            if board[i] == ' ':
                board[i] = PLAYER
                score = alphabeta(board, depth + 1, alpha, beta, True)
                board[i] = ' '

                best_score = min(best_score, score)
                beta = min(beta, best_score)

                if beta <= alpha:
                    break  # CẮT TỈA

        return best_score


# Tìm nước đi tốt nhất
def best_move(board):
    best_score = -math.inf
    move = -1

    for i in range(9):
        if board[i] == ' ':
            board[i] = AI
            score = alphabeta(board, 0, -math.inf, math.inf, False)
            board[i] = ' '

            if score > best_score:
                best_score = score
                move = i

    return move


# -------------------------
# Demo
board = [
    'X','O','X',
    ' ','O',' ',
    ' ',' ','X'
]

print("AI nên đánh vào:", best_move(board))
```

Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.

```python
import math

PLAYER = 'O'
AI = 'X'

# --------------------------
# Kiểm tra thắng trong N×N
# --------------------------
def check_win(board, N, K, symbol):
    # Kiểm tra hàng
    for r in range(N):
        count = 0
        for c in range(N):
            if board[r][c] == symbol:
                count += 1
                if count == K:
                    return True
            else:
                count = 0

    # Kiểm tra cột
    for c in range(N):
        count = 0
        for r in range(N):
            if board[r][c] == symbol:
                count += 1
                if count == K:
                    return True
            else:
                count = 0

    # Kiểm tra chéo chính
    for r in range(N):
        for c in range(N):
            count = 0
            for i in range(K):
                if r+i < N and c+i < N and board[r+i][c+i] == symbol:
                    count += 1
                    if count == K:
                        return True
                else:
                    break

    # Kiểm tra chéo phụ
    for r in range(N):
        for c in range(N):
            count = 0
            for i in range(K):
                if r+i < N and c-i >= 0 and board[r+i][c-i] == symbol:
                    count += 1
                    if count == K:
                        return True
                else:
                    break

    return False


def is_full(board, N):
    for row in board:
        if ' ' in row:
            return False
    return True


# --------------------------
#   MINIMAX (có depth limit)
# --------------------------
def minimax(board, N, K, depth, max_depth, is_maximizing):
    # Kiểm tra trạng thái kết thúc
    if check_win(board, N, K, AI):
        return 10 - depth
    if check_win(board, N, K, PLAYER):
        return depth - 10
    if depth == max_depth or is_full(board, N):
        return 0

    if is_maximizing:
        best = -math.inf
        for r in range(N):
            for c in range(N):
                if board[r][c] == ' ':
                    board[r][c] = AI
                    score = minimax(board, N, K, depth + 1, max_depth, False)
                    board[r][c] = ' '
                    best = max(best, score)
        return best

    else:
        best = math.inf
        for r in range(N):
            for c in range(N):
                if board[r][c] == ' ':
                    board[r][c] = PLAYER
                    score = minimax(board, N, K, depth + 1, max_depth, True)
                    board[r][c] = ' '
                    best = min(best, score)
        return best


# --------------------------
#   Tìm nước đi tốt nhất
# --------------------------
def best_move(board, N, K, max_depth=3):
    best_score = -math.inf
    move = (-1, -1)

    for r in range(N):
        for c in range(N):
            if board[r][c] == ' ':
                board[r][c] = AI
                score = minimax(board, N, K, 0, max_depth, False)
                board[r][c] = ' '

                if score > best_score:
                    best_score = score
                    move = (r, c)

    return move


# --------------------------
#   DEMO TEST 5×5
# --------------------------
N = 5
K = 4  # cần 4 liên tiếp để thắng

board = [
    ['X','O',' ',' ',' '],
    [' ','X','O',' ',' '],
    [' ',' ','X','O',' '],
    [' ',' ',' ',' ',' '],
    [' ',' ',' ',' ',' '],
]

print("AI nên đánh vào:", best_move(board, N, K, max_depth=3))
```
